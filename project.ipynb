Importing the Image Data Generator Libraries 

```python
import tensorflow as tf 
import keras 
from keras.preprocessing.image import ImageDataGenerator
```

Canfiguring The Image Data Generator 

```python
#training images 
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)

#val images 
val_datagen = ImageDataGenerator(rescale=1./255)
```

Applying the image generator for the body 

```python
body_train_generator = train_datagen.flow_from_directory(
        'V:\\WorkSpace\\IBM-Project-23426-1659882722\\Dataset\\Car damage\\body\\training',
        target_size=(150, 150),
        batch_size=32,
        class_mode='categorical')
```

```python
body_validation_generator = val_datagen.flow_from_directory(
        'V:\\WorkSpace\\IBM-Project-23426-1659882722\\Dataset\\Car damage\\body\\validation',
        target_size=(150, 150),
        batch_size=32,
        class_mode='categorical')
```

Applying the image generator for the level 

```python
level_train_generator = train_datagen.flow_from_directory(
        'V:\\WorkSpace\\IBM-Project-23426-1659882722\\Dataset\\Car damage\\level\\training',
        target_size=(150, 150),
        batch_size=32,
        class_mode='categorical')
```

```python
level_validation_generator = val_datagen.flow_from_directory(
        'V:\\WorkSpace\\IBM-Project-23426-1659882722\\Dataset\\Car damage\\level\\validation',
        target_size=(150, 150),
        batch_size=32,
        class_mode='categorical')
```
